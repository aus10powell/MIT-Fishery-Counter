```python
import pandas as pd
import os
import altair as alt
```

```python
DATA_DIR = os.path.join("..", "..", "data")
os.listdir(DATA_DIR)
```

```python
df = pd.read_csv(os.path.join(DATA_DIR, "herring_count_2018.csv"))
df["Date"] = pd.to_datetime(df["Date"])
display(df.head())

df_herring = df[df["Species"] == "Herring"]
df_herring.resample(rule="M", on="Date")["Count"].sum()
```

```python
df1 = pd.read_csv(os.path.join(DATA_DIR, "herring_count_2018.csv"))
df2 = pd.read_csv(os.path.join(DATA_DIR, "herring_count_2017.csv"))
df3 = pd.read_csv(os.path.join(DATA_DIR, "herring_count_2016.csv"))
df4 = pd.read_csv(os.path.join(DATA_DIR, "herring_count_2015.csv"))
```

```python
df2.shape[0] + df1.shape[0] + df3.shape[0] + df4.shape[0]
```

```python
df = pd.concat([df2, df1, df3, df4], 0)
df = df[(df["Species"] == "herring") | (df["Species"] == "Herring")]
df = df.sort_values("Date")
df.to_csv(os.path.join(DATA_DIR, "herring_count_all.csv"), index=False)
```

```python
def plot_count_data(dataframe):
    """Plot data of counts"""
    dataframe["Date"] = pd.to_datetime(dataframe["Date"])
    s = dataframe.resample(rule="D", on="Date")["Count"].sum().reset_index()
    return (
        alt.Chart(s)
        .mark_bar()
        .encode(x="Date", y="Count", tooltip=["Count", "Date"])
        .interactive()
    )


plot_count_data(dataframe=df)
```
